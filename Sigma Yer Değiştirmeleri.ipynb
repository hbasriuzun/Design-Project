```python
from sklearn.preprocessing import StandardScaler

# DataFrame'i NumPy dizisine evirme
data_np = data.values

# StandardScaler kullanarak standardizasyon
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_np)

# Sonucu tekrar DataFrame'e evirme
data_scaled_df = pd.DataFrame(data_scaled, columns=data.columns)

# lk birka satr kontrol etme
print(data_scaled_df.head(10))
```

```python
X = data_scaled_df.iloc[:, :1].values
y = data_scaled_df.iloc[:,1].values
```

```python
X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
rf_reg = RandomForestRegressor(n_estimators = 250 , criterion='squared_error',random_state = 42) 
rf_reg.fit(X_train, y_train)

y_pred = rf_reg.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))

r2_score(y_test, y_pred)

# Use test and show results of test set
y_hat = rf_reg.predict(X)
mae = metrics.mean_absolute_error(y,y_hat)
mse = metrics.mean_squared_error(y,y_hat)
mape = np.mean(np.abs((y - y_hat) / y)) * 100
print ("TEST SET")
print ("========")
print ("MAE:                ", mae)
print ("RMSE:               ", math.sqrt(mse))
print ("MAPE:               ", mape)
print ("r2:                 ", rf_reg.score(X,y))
```

```python
# Gerek ve tahmin edilen deerleri ieren DataFrame
result_df = pd.DataFrame({'Gerek Deerler': y_test, 'Tahmin Edilen Deerler': y_pred})

# Scatter plot izimi
plt.figure(figsize=(20, 12))
sns.scatterplot(x='Gerek Deerler', y='Tahmin Edilen Deerler', data=result_df, palette='viridis', edgecolor='w', alpha=0.7)
plt.title('Gerek vs. Tahmin Edilen Deerler')
plt.xlabel('Gerek Deerler')
plt.ylabel('Tahmin Edilen Deerler')

# Gerek deerlerin younluk grafii
sns.kdeplot(result_df['Gerek Deerler'], label='Gerek Deerler', shade=True)

# Tahmin edilen deerlerin younluk grafii
sns.kdeplot(result_df['Tahmin Edilen Deerler'], label='Tahmin Edilen Deerler', shade=True)

plt.legend()
plt.show() 


df_results = pd.DataFrame({'Predicted': y_pred, 'Actual': y_test})

plt.figure(figsize=(20, 12))
plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.plot(X_test, y_pred, color='red', linewidth=1, label='Predicted Regression Line')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Random Forest Regressor: Predicted vs Actual Values with Regression Line')
plt.legend()
plt.show()
```

### Multi-Layer Perceptron

```python
from keras.models import Sequential
from keras.layers import Dense

mlp = Sequential()
mlp.add(Dense(units=8, kernel_initializer='uniform', activation = 'relu', input_dim = 1))
mlp.add(Dense(units=16, kernel_initializer='uniform', activation = 'relu'))
mlp.add(Dense(units=1, kernel_initializer='uniform', activation = 'linear'))

mlp.compile(optimizer = 'adam', loss =  'mean_squared_error' , metrics = ['mae'] )
mlp.fit(X_train, y_train,epochs=50, batch_size=32, validation_split=0.1)

y_pred = mlp.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

y_pred = mlp.predict(X_test)

# R-squared (r2) deeri
r2 = r2_score(y_test, y_pred)
print(f'R-squared (r2): {r2}')

# Root Mean Squared Error (RMSE) deeri
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f'RMSE: {rmse}')

# Mean Absolute Error (MAE) deeri
mae = mean_absolute_error(y_test, y_pred)
print(f'MAE: {mae}')

# Mean Absolute Percentage Error (MAPE) deeri
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
print(f'MAPE: {mape}')
```

```python
# Modelin eitim srasnda kayp (loss) ve dorulama kayb (validation loss) deerleri
history = mlp.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.1)

# Modelin eitim srecindeki kayp (loss) deerleri
loss = history.history['loss']

# Modelin eitim srecindeki dorulama kayb (validation loss) deerleri
val_loss = history.history['val_loss']

# Eitim ve dorulama kayb grafii izimi
epochs = range(1, len(loss) + 1)

plt.figure(figsize=(20, 12))

plt.plot(epochs, loss, 'bo', label='Loss')
plt.plot(epochs, val_loss, 'r', label='Validation Loss')
plt.title('Loss vs Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```
